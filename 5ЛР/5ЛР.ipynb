```python
# 2.8) Удаление строк с пропущенными значениями
df2 = pd.read_csv('employees.csv')
df2['Gender'] = df2['Gender'].astype(str)
df2 = df2.loc[(df2['Gender'] != "NaN") & (df2['Gender'] != "nan")]
df2
```

```python
# 2.9) создание нового поля вычисленного на основе значений других полей:
# 2.9.1) через выражение на базе имеющихся колонок
df1 = pd.read_csv('employees.csv')
df1['Gender'] = df1['Gender'].astype(str)
gender_list = list(df1['Gender'])
gender_female = []
for i in gender_list:
    if i == "Female":
        gender_female.append("True")
    else:
        gender_female.append("False")
df1['Women?'] = gender_female
df1
```

```python
# 2.9.2) через DataFrame.apply
df1 = pd.read_csv('employees.csv')
df1['Gender'] = df1['Gender'].astype(str)
def function(row):
    if 'Male' in row['Gender']:
        result = (row['Salary'] * row['Bonus %'])/100 + row['Salary']
        return result
df1['Salary*Bonus%'] = df1.apply(function, axis=1)
df1
```

```python
# 2.9.3) через Series.apply
df1 = pd.read_csv('employees.csv')
df1['Gender'] = df1['Gender'].astype(str)
gender_list = list(df1['Gender'])
a = pd.Series(gender_list)
def func(value):
    if value == "Male":
        return True
    return False
a.apply(func)
```

```python
# 2.10) сортировка по какому-либо из полей
# 2.11) вычислить несколько статистик по колонкам.
df1 = pd.read_csv('employees.csv')
mean_salary = df1['Salary'].mean()
min_salary = df1['Salary'].min()
max_salary = df1['Salary'].max()
print("Среднее значение: ", mean_salary)
print("Минимальное значение: ", min_salary)
print("Максимальное значение: ", max_salary)
```

```python
# 2.14) Удалите текущий индекс и создайте новый индекс на базе новой колонки, которая для этого лучше всего подходит
df1 = pd.read_csv('employees.csv')
df1['Gender'] = df1['Gender'].astype(str)
def function(row):
    if 'Male' in row['Gender']:
        result = (row['Salary'] * row['Bonus %'])/100 + row['Salary']
        return result
df1['Salary*Bonus%'] = df1.apply(function, axis=1)
df1.set_index('Salary*Bonus%', inplace= True )
df1
```
